```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.arange(10)
```

```python
sns.set_style("whitegrid")
plt.plot(np.arange(10))
plt.show()
```

```python
sns.set(style="ticks", color_codes=True)
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
plt.show()
```

```python
g = sns.FacetGrid(train_df, col="Sex")
g = g.map(plt.hist, "Age")   #bins=20这个参数控制柱状的数目
plt.show()
```

```python
# 是否获救与船舱等级的关系
g = sns.FacetGrid(train_df, col="Survived")
g = g.map(plt.hist, "Pclass")   #bins=20这个参数控制柱状的数目
plt.show()
```

```python
# 是否获救与登录港口的关系
train_df['Embarked'].fillna('S',inplace=True)        #填充空值
train_df['Embarked'] = train_df['Embarked'].map({'S':0,'C':1,'Q':2}).astype(int)  #把字母类型特征转换为数字
```

```python
g = sns.FacetGrid(train_df, col="Survived")
g = g.map(plt.hist, "Embarked")   #bins=20这个参数控制柱状的数目
plt.show()
```
